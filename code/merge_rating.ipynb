```
from datetime import datetime,timedelta
from math import ceil
from dateutil.parser import parse
```

```

# average_rating
def average_rating(retail,r2):
    st = str(retail.start_time)
    fi = str(retail.start_time+timedelta(minutes=ceil(retail['노출(분)']))) 
    return (r2[st:fi].sum()/ceil(retail['노출(분)']))
retail['rating'] = retail.apply(average_rating,r2=r2,axis=1)
```

```
#max_rating
def max_rating(retail,r2):
    st=str(retail.start_time)
    fi=str(retail.start_time+timedelta(minutes=ceil(retail['노출(분)'])))
    return (r2[st:fi].max())

retail['max_rating']=retail.apply(max_rating,r2=r2,axis=1)
```
